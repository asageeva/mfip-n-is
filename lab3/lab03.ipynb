```python
import numpy as np
```

```python
key = ''
message = ''
alphabet = ''
```

```python
def gamming(k, mes, alp):
    alp = list(alp)
    k = list(k)
    mes = list(mes)
    n = len(alp)
    while len(k) < len(mes):
        k += k
    k = k[:len(mes)]
#     print(alp, k, mes, n)
    mes_i = []
    for i in range(len(mes)):
        for j in range(n):
            if mes[i] == alp[j]:
                mes_i.append(j)
    k_i = []
    for i in range(len(k)):
        for j in range(n):
            if k[i] == alp[j]:
                k_i.append(j)
#     print(mes_i, k_i)
    new_mes =[]
    for i in range(len(mes_i)):
        new_mes.append(alp[(mes_i[i]+k_i[i])%n])
    new_mes = ''.join(new_mes)
    return new_mes
```

```python
new_message = gamming(key, message, alphabet)
```

```python
print(message, "- ")
print(new_message, "-  ")
```
