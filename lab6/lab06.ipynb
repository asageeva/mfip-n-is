## p-метод Полларда

```python
def f(x, n):
    return (x**2 + 5)%n
```

```python
def euclid(a, b): 
    while a != 0 and b != 0:
        if a >= b:
            a %= b
        else:
            b %= a
    return a or b
```

```python
def pollard(n, c):
    a = c
    b = c
    while True:
        a = f(a,n)%n
        b = f(f(b,n),n)%n
        d = euclid(abs(a-b), n)
        print(a, b, d)
        if 1 < d < n:
            p = d
            return p
        elif d == n:
            return "Делитель не найден"
```
