```python
import numpy as np
```

```python
k = 3 #  
message = "venivedivici"
alphabet = "abcdefghigklmnopqrstuvwxyz"
```

```python
def caesar(k, word, alphabet):
    m = len(alphabet)
    alphabet = list(alphabet)
    word = list(word)
    new_alphabet = []
    for i in range(m):
        new_alphabet.append(alphabet[(i+k)%m])
    new_word = []
    for i in range(len(word)):
        for j in range(m):
            if word[i] == alphabet[j]:
                new_word.append(new_alphabet[j])
    new_word = ''.join(new_word)
    new_alphabet = ''.join(new_alphabet)
    return new_word, new_alphabet
```

```python
new_message, new_alphabet = caesar(k, message, alphabet)
```

```python
print(k, "- ")
print(alphabet, "- ")
print(new_alphabet, "-  ")
print(message, "- ")
print(new_message, "-  ")
```

```python
alphabet = ' '
message = '  '
```

```python
def atbash(alphabet, word):
    m = len(alphabet)
    alphabet = list(alphabet)
    new_alphabet = []
    for i in range(m):
        new_alphabet.append(alphabet[m-i-1])
    word = list(word)
    new_word = []
    for i in range(len(word)):
        for j in range(m):
            if word[i] == alphabet[j]:
                new_word.append(new_alphabet[j])
    new_word = ''.join(new_word)
    new_alphabet = ''.join(new_alphabet)
    return new_alphabet, new_word
```

```python
new_alphabet, new_message = atbash(alphabet, message)
```

```python
print(alphabet, "- ")
print(new_alphabet, "-  ")
print(message, "- ")
print(new_message, "-  ")
```
